```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
graffiti_file = "311-service-requests-graffiti-removal.csv"
graffiti_df = pd.read_csv(graffiti_file)
graffiti_df.head()
```

```python
lights_all_file = "311-service-requests-street-lights-all-out.csv"
lights_all_df = pd.read_csv(lights_all_file)
lights_all_df.head()
```

```python
lights_one_file = "311-service-requests-street-lights-one-out.csv"
lights_one_df = pd.read_csv(lights_one_file)
lights_one_df.head()
```

```python
alley_lights_file = "311-service-requests-alley-lights-out.csv"
alley_lights_df = pd.read_csv(alley_lights_file)
alley_lights_df.head()
```

```python
graffiti_cols = ["Creation Date", "Service Request Number", "Type of Service Request", "ZIP Code", "Ward", "Police District", "Community Area", "What Type of Surface is the Graffiti on?" ]
graffiti_transformed= graffiti_df[graffiti_cols].copy()

# Rename the column headers
graffiti_transformed = graffiti_transformed.rename(columns={"Creation Date": "creation_date",
                                                          "Service Request Number": "service_request_number",
                                                          "Type of Service Request": "graffiti_request",
                                                          "ZIP Code": "ZIP",
                                                          "Ward": "ward",
                                                          "Police District": "police_district",
                                                          "Community Area": "community_area",
                                                          "What Type of Surface is the Graffiti on?": "graffiti_location"})

# Clean the data by dropping duplicates and setting the index
graffiti_transformed.drop_duplicates("service_request_number", inplace=True)
graffiti_transformed.set_index("service_request_number", inplace=True)

graffiti_transformed.head()
```

```python
lights_all_cols = ["Creation Date", "Service Request Number", "Type of Service Request", "ZIP Code", "Ward", "Police District", "Community Area"]
lights_all_transformed= lights_all_df[lights_all_cols].copy()

# Rename the column headers
lights_all_transformed = lights_all_transformed.rename(columns={"Creation Date": "creation_date",
                                                          "Service Request Number": "service_request_number",
                                                          "Type of Service Request": "st_lights_all_request",
                                                          "ZIP Code": "ZIP",
                                                          "Ward": "ward",
                                                          "Police District": "police_district",
                                                          "Community Area": "community_area"})

# Clean the data by dropping duplicates and setting the index
lights_all_transformed.drop_duplicates("service_request_number", inplace=True)
lights_all_transformed.set_index("service_request_number", inplace=True)

lights_all_transformed.head()
```

```python
lights_one_cols = ["Creation Date", "Service Request Number", "Type of Service Request", "ZIP Code", "Ward", "Police District", "Community Area"]
lights_one_transformed= lights_one_df[lights_all_cols].copy()

# Rename the column headers
lights_one_transformed = lights_one_transformed.rename(columns={"Creation Date": "creation_date",
                                                          "Service Request Number": "service_request_number",
                                                          "Type of Service Request": "st_lights_one_request",
                                                          "ZIP Code": "ZIP",
                                                          "Ward": "ward",
                                                          "Police District": "police_district",
                                                          "Community Area": "community_area"})

# Clean the data by dropping duplicates and setting the index
lights_one_transformed.drop_duplicates("service_request_number", inplace=True)
lights_one_transformed.set_index("service_request_number", inplace=True)

lights_one_transformed.head()
```

```python
alley_lights_cols = ["Creation Date", "Service Request Number", "Type of Service Request", "ZIP Code", "Ward", "Police District", "Community Area"]
alley_lights_transformed= alley_lights_df[alley_lights_cols].copy()

# Rename the column headers
alley_lights_transformed = alley_lights_transformed.rename(columns={"Creation Date": "creation_date",
                                                          "Service Request Number": "service_request_number",
                                                          "Type of Service Request": "alley_lights_request",
                                                          "ZIP Code": "ZIP",
                                                          "Ward": "ward",
                                                          "Police District": "police_district",
                                                          "Community Area": "community_area"})

# Clean the data by dropping duplicates and setting the index
alley_lights_transformed.drop_duplicates("service_request_number", inplace=True)
alley_lights_transformed.set_index("service_request_number", inplace=True)

lights_one_transformed.head()
```
